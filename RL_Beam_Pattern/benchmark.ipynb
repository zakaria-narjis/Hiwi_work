```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def plot(beam_id):

    plt.figure(figsize=(12, 6))
    plt.title(f'Best beamforming gain')
    plt.xlabel('Beam_framing_gain')  

    plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
    for algo in ['ddpg','myddpg','sac']:
        beam_path = f'BF_40K_{algo}/beams_gain_records/beam_{beam_id}_gain_records.npy'
        gain_record = np.sqrt(np.load(beam_path))
        plt.plot(gain_record, label=f'Beam_{beam_id}_{algo}')

    plt.plot([np.sqrt(231.17094)]*40000,linestyle='dashed',color='black')
    plt.legend(loc="lower right")  
    plt.show() 
```
