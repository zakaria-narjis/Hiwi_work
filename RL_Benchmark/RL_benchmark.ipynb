```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
algos=['DQN','DDQN','DuelingDQN','Rainbow','Categorical_DQN','TRPO','A2C','PPO']
```

```python
 data = pd.read_csv('/content/scores/scores_DQN.txt', header = 0,delimiter='\t')
```

```python
algos_results = pd.DataFrame({'steps':[20000,40000,80000,100000]})
for algo in algos:
  data = pd.read_csv(f'/content/scores/scores_{algo}.txt', header = 0,delimiter='\t')
  algos_results[algo] = data['mean']
```

```python

plt.figure(figsize=(10,6))

for algo in algos:
  plt.plot(algos_results['steps'],algos_results[algo],label=algo)
plt.legend(loc="upper left")
plt.title("RL_Benchmark_On_Cifar10")
plt.xlabel("steps")
plt.ylabel("Accuracy_on_test_dataset")

plt.show()
```
